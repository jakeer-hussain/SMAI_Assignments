```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
file_path = "train_data.csv"  # Update if needed
df = pd.read_csv(file_path)

# Drop Address column (not needed for correlation)
df_numeric = df.drop(columns=["Address"])

# Plot correlation matrix
plt.figure(figsize=(12, 8))
corr_matrix = df_numeric.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix")
plt.show()

# Violin plot for each feature against FLAG
plt.figure(figsize=(15, 10))
for i, column in enumerate(df_numeric.columns[1:], 1):  # Exclude FLAG from x-axis
    plt.subplot(4, 4, i)
    sns.violinplot(x=df_numeric["FLAG"], y=df_numeric[column])
    plt.title(column)
    plt.xlabel("FLAG")
    plt.ylabel(column)

plt.tight_layout()
plt.show()
```
