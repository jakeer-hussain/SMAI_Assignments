```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("Food_Delivery_Times.csv")

# Generate FutureWarnings by using deprecated parameters
plt.figure(figsize=(15, 10))

categorical_features = ['Weather', 'Traffic_Level', 'Time_of_Day', 'Vehicle_Type', 'Courier_Experience_yrs']
for feature in categorical_features:
    plt.figure(figsize=(8, 5))

    # FutureWarning: Using `palette` without `hue`
    sns.countplot(data=df, x=feature, palette='Set2')

    plt.title(f'Bar Plot of {feature}')
    plt.show()


continuous_features = ['Distance_km', 'Preparation_Time', 'Delivery_Time_min']
for feature in continuous_features:
    plt.figure(figsize=(8, 5))

    # FutureWarning: `saturation` parameter will be removed in future versions
    sns.violinplot(data=df, x=feature, palette='Set2', saturation=0.75)

    plt.title(f'Violin Plot of {feature}')
    plt.xticks(rotation=45)
    plt.show()
```
